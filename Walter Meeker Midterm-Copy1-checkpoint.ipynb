The height feature is in wrong format, want to convert to inches.

```python
 # import regex to conver the given height
r = re.compile(r"([0-9]+)-([0-9]+)")
def get_inches(el):
    '''this function gets the height in feet as first group and heigh in inches 
    as second group and calcs height in inches'''
    m = r.match(el)
    if m == None:
        return float('NaN')
    else:
        return int(m.group(1))*12 + float(m.group(2))
# test that the function works
get_inches('6-10')
```

need to convert height to inches, define function to do so and apply to combined_df_final

```python
# apply the function to height column
combined_df_final['height'] = combined_df_final['height'].apply(lambda x: get_inches(x))
combined_df_final.head()
```

## Start With Linear Regression and See What turns out

```python
# dictionary that will append results of multiple models to
results = {}

# instantiate and fit LR model
lr = LinearRegression(fit_intercept= False)
lr.fit(X_train, y_train) # don't fit intercept because didn't drop one of the dummies

# predict train and test
y_pred_test_lr = lr.predict(X_test)
y_pred_train_lr = lr.predict(X_train)

# calc mse and append to dictionary
rmse_test_lr = sqrt(mean_squared_error(y_test.values, y_pred_test_lr))
rmse_train_lr = sqrt(mean_squared_error(y_train.values, y_pred_train_lr))

# append results to dictionary
results['rmse test lr'] = rmse_test_lr
results['rmse train lr'] = rmse_train_lr

# plot the results on test
plt.plot(y_test.values, label = "Actual", alpha= 0.5)
plt.plot(y_pred_test_lr, label = "Prediction", alpha= 0.7)
plt.title('Actual vs Prediction of the Test - Linear Regression')
plt.legend()
plt.savefig('Actual vs Prediction of the Test - Linear Regression.pdf')
plt.show()

# plot the results on train
plt.plot(y_train.values, label = "Actual", alpha= 0.5)
plt.plot(y_pred_train_lr, label = "Prediction", alpha= 0.7)
plt.title('Actual vs Prediction of the Train - Linear Regression')
plt.legend()
plt.savefig('Actual vs Prediction of the Train - Linear Regression.pdf')
plt.show()
```
